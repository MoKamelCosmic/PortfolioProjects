Importing the standard libraries and exploring the data....

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

seems like we want to predict the diagnosis, to check if a tumor is benign or malignant, lets explore the distribution in the main data frame

```python
sns.countplot(data=df,x='diagnosis')
```

sounds like the classes are well balanced, so accuracy would be a good metric to use to evaluate our model
lets convert the diagnosis columns into 0s and 1s to feed it into the model later on, wel make a new dataframe with the features ready for the model

```python
df2 = pd.get_dummies(data=df,columns=['diagnosis'],dtype=int)
```

```python
#drop the original diagnosis column
df2 = df2.drop(['diagnosis_M'],axis = 1)
```

```python
df2.info()
```

```python
#Drop null values
df2 = df2.drop(['Unnamed: 32'],axis=1)
```

```python
#Drop id column, it is useless and can alter the model based on irrelevant information
df2= df2.drop(['id'],axis=1)
```

lets check the correlation between each of the features and the diagnosis, we will use a bar and a heatmap as they show it in a neat way

```python
df2.corr(numeric_only=True)['diagnosis_B'][:-1].sort_values().plot(kind='bar')
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(df2.corr())
```

lets now do our train test split, then we will test our model 3 times: 
1. without callbacks or dropping layers
2. with callbacks
3. with callbacks and dropping

```python
df2.columns
```

```python
X = df2.drop(['diagnosis_B'],axis=1).values
y = df2['diagnosis_B'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

Then we scale the data:

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

import Dense and Dropout and lets start building the model, we will choose a very high number of epochs to display overfitting, then we will use callbacks and dropouts to deal with overfitting

```python
from tensorflow.keras.layers import Dense,Dropout
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))

model.add(Dense(15,activation='relu'))
# BINARY CLASSIFICATION
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,epochs=1200,validation_data=(X_test,y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

now that is the epitome of overfitting! lets see if early stopping callback does anything

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=20)
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

This is defnitely much better, lets try predicting on our test set and evaluate using the classification report and confusion matrix

```python
predictions = (model.predict(X_test) > 0.5).astype("int32")
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
